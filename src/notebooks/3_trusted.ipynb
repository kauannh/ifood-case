```python
CATALOG_NAME = "ifood_case"
SOURCE_SCHEMA_NAME = "raw"
TARGET_SCHEMA_NAME = "trusted"
TABLE_NAME = "ny_taxi_trip_yellow"
```

```python
spark.sql(f"""
SELECT CAST(VendorID AS LONG) AS vendor_id,
       CAST(passenger_count AS DOUBLE) AS passenger_count,
       CAST(total_amount AS DOUBLE) AS total_amount,
       CAST(tpep_pickup_datetime AS TIMESTAMP) AS tpep_pickup_datetime,
       CAST(tpep_dropoff_datetime AS TIMESTAMP) AS tpep_dropoff_datetime,
       year,
       month
FROM {CATALOG_NAME}.{SOURCE_SCHEMA_NAME}.{TABLE_NAME}
""").show()
```
