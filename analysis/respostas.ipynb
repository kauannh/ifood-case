```python
%sql
-- Qual a média de valor total (total\_amount) recebido em um mês
-- considerando todos os yellow táxis da frota?

SELECT
    date_format(date_trunc('MM', tpep_pickup_datetime), 'yyyy-MM') as tpep_pickup_month,
    AVG(total_amount) AS avg_total_amount
FROM ifood_case.trusted.ny_taxi_trip_yellow
-- FILTRO OPCIONAL PARA OBTER APENAS CORRIDAS DE JAN/2023 A MAI/2023
--WHERE tpep_pickup_datetime BETWEEN '2023-01-01' AND '2023-05-31'
GROUP BY date_format(date_trunc('MM', tpep_pickup_datetime), 'yyyy-MM')
ORDER BY 1;
```

```python
%sql
-- Qual a média de passageiros (passenger\_count) por cada hora do dia
-- que pegaram táxi no mês de maio considerando todos os táxis da frota?

WITH total_passengers AS (
    SELECT tpep_pickup_datetime AS pickup_datetime, passenger_count
    FROM ifood_case.trusted.ny_taxi_trip_yellow
    UNION ALL
    SELECT lpep_pickup_datetime AS pickup_datetime, passenger_count
    FROM ifood_case.trusted.ny_taxi_trip_green
)
SELECT
    date_format(pickup_datetime, 'HH') AS pickup_hour,
    AVG(passenger_count) AS avg_passenger_count
FROM total_passengers
WHERE pickup_datetime >= '2023-05-01'
  AND pickup_datetime <  '2023-06-01'
  -- FILTRO PARA REMOVER CORRIDAS SEM PASSAGEIROS
  AND passenger_count > 0
GROUP BY date_format(pickup_datetime, 'HH')
ORDER BY 1;
```
